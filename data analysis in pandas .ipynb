```python
df3.groupby(['race_length','athletes_gender'])['athletes_average_speed'].mean()
```

```python
#what age groups are the best in the 50m race (20 + race min)
```

```python
df3.query('race_length == "50mi"') \
   .groupby('athletes_age')['athletes_average_speed'] \
   .agg(['mean', 'count']) \
   .sort_values('mean', ascending=False) \
   .query('count > 19')
```

```python
#what age groups are the best in the 50m race (10 + race min)
```

```python
df3.query('race_length == "50mi"') \
   .groupby('athletes_age')['athletes_average_speed'] \
   .agg(['mean', 'count']) \
   .sort_values('mean', ascending=True) \
   .query('count > 9')
```
